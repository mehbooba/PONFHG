```python
import re
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["figure.figsize"] = (10,6)
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 14}

matplotlib.rc('font', **font)

Datasetsize=['Precision','Recall','NDCG','HR']
```

```python
PO_NFHG=[.125,.156,.27,.032]
topn=[5,10,15,20,25]
F1score=[]
Precision= [0.032, 0.03, 0.0283,0.028,0.028]
recall= [0.15,0.152,0.152,0.153,0.156]
#F1score= [0.345, 0.565,0.672,0.706,0.711]
HR= [.27,.27,.279,.2795,.28]
NDCG= [0.124,0.125, 0.1251, 0.125,0.126]
F1score=[]
for precision, recall in zip(Precision, recall):
    if precision + recall == 0:
        f1 = 0  # Handle division by zero
    else:
    	f1 = 2 * (precision * recall) / (precision + recall)
    F1score.append(f1)
X_axis = np.arange(len(topn))
plt.bar(X_axis - 0.1,Precision , 0.2, label = 'Precision@N',color='blue')
plt.bar(X_axis + 0.1,recall , 0.2, label = 'Recall@N',color="red")
plt.bar(X_axis - 0.2,F1score , 0.2, label = 'F1Score@N',color="green")
plt.bar(X_axis + 0.2,HR, 0.2, label = 'HR@N',color='brown')
plt.bar(X_axis - 0.3,NDCG , 0.2, label = 'NDCG@N',color="orange")
#plt.bar(X_axis + 0.3, RMSE, 0.2, label = 'RMSE@N',color="brown")
plt.xlabel(" VALUES OF N", weight="bold")
plt.xticks(X_axis, topn)
#plt.xticks(rotation = 45, fontsize=12,fontname='Times')
plt.ylabel("EVALUATION PARAMETERS",weight="bold")
plt.legend(ncol=6, loc="upper right",fontsize="xx-small")
plt.savefig("topns_partial.jpg", dpi=300,bbox_inches="tight")
plt.show()
```
